```python
import seaborn as sns 
import pandas as pd
tips = sns.load_dataset("tips")
tips.head()
```

## 문제
- total_bill이 20보다 큰 행들만 필터링하세요.

```python
tips.loc[tips['total_bill'] > 20, :].reset_index(drop=True)
```

- total_bill이 20보다 크고 tip이 3보다 큰 행들만 필터링하세요.

```python
tips.loc[(tips['total_bill'] > 20) & (tips['tip'] > 3)].reset_index(drop=True)
```

## 문제 
- Sun, Sat만 출력

```python
tips['day'].unique()
```

```python
target_days = ['Sat', 'Sun']
tips.loc[tips['day'].isin(target_days)].reset_index(drop=True)
```

## 문제
- New가 들어간 도시만 출력
- contains 메서드 활용

```python
import pandas as pd

data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'city': ['New York', 'Los Angeles', 'New Yorker', 'Chicago', 'New City'],
    'age': [25, 30, 35, 40, 45]
}

df = pd.DataFrame(data)
df
```

```python
df.loc[df['city'].str.contains('New')].reset_index(drop=True)
```

```python
# GroupBy
# dataset.groupby(column_name).aggregation_function
# 데이터 형태가 바뀜 + 추가적인 조작
```

```python
# 남자와 여자 중 어느 쪽이 평균적으로 팁을 더 많이 줄까?

tips.groupby('sex', observed=True)['tip'].mean()
```

```python
tips.groupby('sex', observed=True)['tip'].mean().reset_index()
```

```python
tips.groupby(['sex', 'day', 'smoker'], observed=True)['tip'].mean()
```

```python
tips.groupby('sex', observed=True)['tip'].agg(['mean', 'std', 'max']).reset_index() # 여러 개의 집계 함수를 한 번에
```

```python
tips.groupby('sex', observed=True).size().reset_index() # 각 행의 갯수
```
